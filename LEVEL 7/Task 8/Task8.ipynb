```python
import numpy as np
import matplotlib.pyplot as plt
n_trials = 10       
p_success = 0.5     
num_experiments = 1000 
data = np.random.binomial(n=n_trials, p=p_success, size=num_experiments)
simulated_mean = np.mean(data)
plt.hist(data, bins=range(n_trials + 2), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5, 1000 Experiments)')
plt.xticks(range(n_trials + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print(f"Simulated Mean: {simulated_mean:.4f}")
print(f"Expected Mean: {n_trials * p_success:.4f}")
print(f"Interpretation: The simulated mean is close to the expected binomial mean (n × p = 5), confirming the distribution behavior.")
```
