```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')
print("First 5 rows:\n", df.head())
print("\nDataset Info:\n")
df.info()
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
print("\n--- Measures of Central Tendency ---")
for col in numerical_cols:
    print(f"\nColumn: {col}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Mode: {df[col].mode().values}")
print("\n--- Mean values grouped by Survived ---")
print(df.groupby('survived')[numerical_cols].mean())
categorical_cols = ['sex', 'pclass', 'embarked', 'alone']
for col in categorical_cols:
    sns.countplot(data=df, x=col, hue='survived')
    plt.title(f"Survival Count by {col}")
    plt.legend(title='Survived', labels=['No', 'Yes'])
    plt.show()
for col in ['age', 'fare']:
    sns.boxplot(data=df, x='survived', y=col)
    plt.title(f"{col.capitalize()} Distribution by Survival")
    plt.xticks([0,1], ['No', 'Yes'])
    plt.show()
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
